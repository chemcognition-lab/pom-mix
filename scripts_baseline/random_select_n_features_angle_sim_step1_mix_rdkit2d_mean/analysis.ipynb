```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
for i in range(0, 5):
    df = pd.read_csv(f'select_n_features_cv{i}.csv')

    plt.scatter(data=df, x="n_features", y="rmse_avg")
    plt.errorbar(
        df["n_features"],
        df["rmse_avg"],
        yerr=df["rmse_std"],
        fmt="o",
    )
    plt.title(f"RMSE vs. Number of features (CV={i})")
    plt.xlabel("Number of features")
    plt.ylabel("Mean RMSE")
    plt.legend()
```
