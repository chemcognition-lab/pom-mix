```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('select_n_features.csv')

plt.scatter(data=df, x="n_features", y="rmse_avg")
plt.errorbar(
    df["n_features"],
    df["rmse_avg"],
    yerr=df["rmse_std"],
    fmt="o",
)
plt.xlabel("Number of features")
plt.ylabel("Mean RMSE")
```

```python
import scipy.stats as stats
zscore = -stats.zscore(df["rmse_avg"])
# set all negative values to 0
zscore = [max(0, x) for x in zscore]
# get index of positive features
positive_features = [i for i, x in enumerate(zscore) if x > 0]
positive_features
# df["score"] = df["rmse_avg"].apply(lambda x: max(0, -stats.zscore([x])[0]))
# df
```
