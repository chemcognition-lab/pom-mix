```python
import pandas as pd
import numpy as np
```

```python
molecules = pd.read_csv("molecules.csv")
molecules.drop_duplicates(inplace=True)
stimuli = pd.read_csv("stimuli.csv")
tables1 = pd.read_csv("tableS1.csv")
tables2 = pd.read_csv("tableS2.csv")
```

# Expt 1

```python
tables1 = pd.read_csv("tableS1.csv")
expt1_mixture_definitions = tables1[(tables1['Type'] == 'mixture') & (tables1['Experiment'] == 'Exp1')].copy()
expt1_mixture_definitions.dropna(axis=1, how='all', inplace=True)
for i in range(1,11):
    expt1_mixture_definitions = pd.merge(expt1_mixture_definitions, molecules[['CID', 'IsomericSMILES']], left_on=f'Mix Component {i}', right_on='CID', how='left')
    expt1_mixture_definitions[f'Mix Component {i}'] = expt1_mixture_definitions['IsomericSMILES']
    expt1_mixture_definitions.drop(columns=['CID', 'IsomericSMILES'], inplace=True)
desired_list = ['ID In Exp 1'] + [f'Mix Component {i}' for i in range(1,11)]
expt1_mixture_definitions = expt1_mixture_definitions[desired_list]
desired_column_names = ['Mixture Label'] + [f'smi_{i}' for i in range(10)]
expt1_mixture_definitions.columns = desired_column_names
for i in range(10, 43):
    expt1_mixture_definitions[f'smi_{i}'] = np.nan
expt1_mixture_definitions['Dataset'] = 'Ravia 1'
ordered_cols = ["Dataset"] + expt1_mixture_definitions.columns[:-1].to_list()
expt1_mixture_definitions = expt1_mixture_definitions[ordered_cols]
```

```python
expt1_behavior = pd.read_csv("behavior_2.csv")
expt1_behavior = expt1_behavior[(expt1_behavior['Stimulus 1'] <= 270) & (expt1_behavior['Stimulus 2'] <= 270)]
expt1_behavior['Stimulus 1'] -= 256
expt1_behavior['Stimulus 2'] -= 256
expt1_behavior['Dataset'] = 'Ravia 1'
expt1_behavior.columns = ['Mixture 1', 'Mixture 2', 'Experimental Values', 'Dataset']
expt1_behavior['Experimental Values'] = (100-expt1_behavior['Experimental Values'])/100
ordered_cols = ["Dataset"] + expt1_behavior.columns[:-1].to_list()
expt1_behavior = expt1_behavior[ordered_cols]
```

# Expt 2

```python
tables1 = pd.read_csv("tableS1.csv")
expt2_mixture_definitions = tables1[(tables1['Type'] == 'mixture') & (tables1['Experiment'] == 'Exp2')].copy()
expt2_mixture_definitions.dropna(axis=1, how='all', inplace=True)
for i in range(1,11):
    expt2_mixture_definitions = pd.merge(expt2_mixture_definitions, molecules[['CID', 'IsomericSMILES']], left_on=f'Mix Component {i}', right_on='CID', how='left')
    expt2_mixture_definitions[f'Mix Component {i}'] = expt2_mixture_definitions['IsomericSMILES']
    expt2_mixture_definitions.drop(columns=['CID', 'IsomericSMILES'], inplace=True)
desired_list = ['ID In Exp 1'] + [f'Mix Component {i}' for i in range(1,11)]
expt2_mixture_definitions = expt2_mixture_definitions[desired_list]
desired_column_names = ['Mixture Label'] + [f'smi_{i}' for i in range(10)]
expt2_mixture_definitions.columns = desired_column_names
for i in range(10, 43):
    expt2_mixture_definitions[f'smi_{i}'] = np.nan
expt2_mixture_definitions['Dataset'] = 'Ravia 2'
ordered_cols = ["Dataset"] + expt2_mixture_definitions.columns[:-1].to_list()
expt2_mixture_definitions = expt2_mixture_definitions[ordered_cols]
```

```python
expt2_behavior = pd.read_csv("behavior_2.csv")
expt2_behavior = expt2_behavior[(expt2_behavior['Stimulus 1'] > 270) & (expt2_behavior['Stimulus 2'] > 270)]
expt2_behavior['Stimulus 1'] -= 270
expt2_behavior['Stimulus 2'] -= 270
expt2_behavior['Dataset'] = 'Ravia 2'
expt2_behavior.columns = ['Mixture 1', 'Mixture 2', 'Experimental Values', 'Dataset']
expt2_behavior['Experimental Values'] = (100-expt2_behavior['Experimental Values'])/100
ordered_cols = ["Dataset"] + expt2_behavior.columns[:-1].to_list()
expt2_behavior = expt2_behavior[ordered_cols]
```

```python
expt5_raw_behavior = pd.read_csv("tableS3.csv")
expt5_raw_behavior = expt5_raw_behavior[expt5_raw_behavior['Experiment'] == 'Exp4']

pairs = set(list(zip(expt5_raw_behavior['Mixture1'], expt5_raw_behavior['Mixture2'])))

# Create an empty list to store results
results_list = []

# Iterate through all pairs
for stim1, stim2 in pairs:
    # Subset the dataframe for this pair
    subset = expt5_raw_behavior[(expt5_raw_behavior['Mixture1'] == stim1) & (expt5_raw_behavior['Mixture2'] == stim2)]

    print(f"{stim1, stim2}: {len(subset)}")
    
    # Calculate the ratio if the subset is not empty
    if len(subset) > 0:
        ratio = subset['IsCorrect'].sum() / len(subset)
        
        # Add the result to the results list
        results_list.append({
            'Mixture 1': stim1,
            'Mixture 2': stim2,
            'Experimental Values': ratio
        })

# Create the results dataframe from the list
expt5_behavior = pd.DataFrame(results_list)
```
