```python
import pandas as pd
import numpy as np
import os
import sys
from pathlib import Path

script_dir = Path(os.path.abspath(""))
base_dir = Path(*script_dir.parts[:-2])
sys.path.append( str(base_dir / 'src/') )
DATASET_DIR = base_dir / 'datasets'
```

```python
mean_df = pd.read_csv(DATASET_DIR / "mixtures/mixture_rdkit_mean_definitions_clean.csv")
pna_df = pd.read_csv(DATASET_DIR / "mixtures/mixture_rdkit_definitions_clean.csv")
mix_df = pd.read_csv(DATASET_DIR / "mixtures/mixtures_combined.csv")
smi_df = pd.read_csv(DATASET_DIR / "mixtures/mixture_smi_definitions_clean.csv")

smi_df['length'] = smi_df.apply(lambda row: len(row[smi_df.filter(like='smi_').columns].dropna()), axis=1)
```

```python
# In smi_df, check if each row is unique compared to other rows based on values from columns 2 onwards  
set_list = {}
for i, row in smi_df.iterrows():
    curr_set = frozenset(row[2:].dropna())
    if curr_set in set_list.keys():
        print(f"Duplicate row found between {set_list[curr_set]} and [{row['Dataset']}, {row['Mixture Label']}]")
    else:
        set_list[curr_set] = [row['Dataset'], row['Mixture Label']]
```

```python
# In smi_df, make the first instance of each unique row the canonical representation. 
# Make this column duplicate as False for the first instance of the row, and define all other instances of the row by the specific Dataset and Mixture Label of the first instance.
canonical_smi_df = smi_df.copy()
canonical_smi_df['Duplicate'] = False
canonical_smi_df['Duplicate Of'] = np.nan

canonical_pna_df = pna_df.copy()
canonical_pna_df['Duplicate'] = False
canonical_pna_df['Duplicate Of'] = np.nan

canonical_mean_df = mean_df.copy()
canonical_mean_df['Duplicate'] = False
canonical_mean_df['Duplicate Of'] = np.nan
for i, row in smi_df.iterrows():
    curr_set = frozenset(row[2:].dropna())
    if curr_set in set_list.keys():
        if set_list[curr_set] != [row['Dataset'], row['Mixture Label']]:
            for dataframe in [canonical_smi_df, canonical_pna_df, canonical_mean_df]:
                dataframe.loc[i, 'Duplicate'] = True
                dataframe.loc[i, 'Duplicate Of'] = str(set_list[curr_set])
            

# Rearrange columns to bring the Duplicate Columns to the front
canonical_smi_df = canonical_smi_df[['Dataset', 'Mixture Label', 'length', 'Duplicate', 'Duplicate Of'] + list(canonical_smi_df.columns[2:-2])]
canonical_pna_df = canonical_pna_df[['Dataset', 'Mixture Label', 'Duplicate', 'Duplicate Of'] + list(canonical_pna_df.columns[2:-2])]
canonical_mean_df = canonical_mean_df[['Dataset', 'Mixture Label', 'Duplicate', 'Duplicate Of'] + list(canonical_mean_df.columns[2:-2])]
```

```python
canonical_smi_df.to_csv("mixture_smi_definitions_clean_found_duplicates.csv", index=False)
canonical_pna_df.to_csv("mixture_rdkit_definitions_clean_found_duplicates.csv", index=False)
canonical_mean_df.to_csv("mixture_rdkit_mean_definitions_clean_found_duplicates.csv", index=False)
```
